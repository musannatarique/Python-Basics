```python
#Separating THe Legit ANd Fraud Transactions
Legit = Credit_Card_Dataset[Credit_Card_Dataset.Class == 0]
Fraud = Credit_Card_Dataset[Credit_Card_Dataset.Class == 1]
```

```python
print(Legit.shape)
print(Fraud.shape)
```

Building a sample Dataset containing similar distribution of Legit and Fraud


Number of Fraud Transactions --> 492

```python
#Under Sampling
Legit_Sample = Legit.sample(n=492)
print (Legit_Sample.shape)
```

```python
#Concatenate THe 2 DataFrames
New_Dataset = pd.concat([Legit_Sample, Fraud], axis= 0) #axis 0 means "Row Wise"
New_Dataset.head()
```
