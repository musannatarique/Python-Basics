```python
#Aggregate Statistics (Groupby)
# Group the DataFrame by unique combinations of 'Type 1' and 'Type 2'.
# The grouping creates subsets of rows where the values of 'Type 1' and 'Type 2' are the same.
df.groupby(['Type 1', 'Type 2'])

# Use `.size()` to count the number of rows within each group.
# Unlike `.count()`, `.size()` counts all rows, including those with NaN values in other columns.
# The result is a Series with the index as the group combinations and values as the row counts.
df.groupby(['Type 1', 'Type 2']).size()
```

```python
#Working with large amounts of data
# Initialize an empty DataFrame with the same columns as the original dataset.
# This DataFrame will store the final results after processing each chunk.
new_df = pd.DataFrame(columns=df.columns)

# Read the CSV file in chunks of 5 rows at a time using the 'chunksize' parameter.
# This approach is memory-efficient for large datasets.
for df in pd.read_csv(r"D:\000\Code\Py\ML\Data Sheets\Poki_modified.csv", chunksize=5):
    
    # Group the current chunk by the 'Type 1' column and count non-null values in each group.
    # The result is a DataFrame with each unique 'Type 1' value as an index.
    results = df.groupby(['Type 1']).count()
    
    # Append the current chunk's results to 'new_df' using pd.concat().
    # pd.concat() combines DataFrames vertically (row-wise) by default.
    new_df = pd.concat([new_df, results])
    print(new_df)

# At this point, 'new_df' contains the grouped count results from all chunks combined.
    '''This approach is useful when working with large datasets that don't fit into memory'''
```
