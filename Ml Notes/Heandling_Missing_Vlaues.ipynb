```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

**Central Tendencies** 
* Mean
* Median
* Mode

```python
#Analyse The Distribution of Data In Salary 
sns.displot(Dataset.salary)
plt.show()
```

```python
#Analyse The Distribution of Data In Salary (Improve) 
# Set the figure size
plt.figure(figsize=(8, 6))

# Histogram with KDE line
sns.histplot(Dataset['salary'], bins=20, kde=True, color='skyblue', stat='density', alpha=0.7, edgecolor='black')

# Set labels and title
plt.xlabel('Salary')
plt.ylabel('Density')
plt.title('Salary Distribution with KDE Line')

# Display the plot
plt.show()
```

```python
#Replace The Missing Vlaues With --Median-- Vlaue
Dataset['salary'].fillna(Dataset['salary'].median(),inplace=True)
#DataFrame.fillna(value, method, axis, inplace, limit, downcast)
```
